```python
import algotrading
def test(stock_name, days=200):
    st = algotrading.stock.Stock(stock_name)
    st.read_data(days)
    st.generate_more_data()
    st.df[["Close", "ATR","Momentum", "MFI", "Williams_%R"]].plot(subplots=True)
    return st.df
```
