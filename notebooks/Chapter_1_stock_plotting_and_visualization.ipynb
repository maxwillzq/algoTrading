```python
# cpi
fred = algotrading.fred.Fred('CPIAUCSL')
fred.read_data(days=2 * 365)
cpi = fred.df / fred.df.shift(12) - 1.0
cpi = cpi * 100.0
```

```python
cpi.plot(grid=True)
```
